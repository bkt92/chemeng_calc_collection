```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=np.linspace(70+273.15,150+273.15,100)
```

```python
k1=6.147*np.exp(-5673.35/T)
k2=0.683*np.exp(-5268.02/T)
```

```python
fig, ax = plt.subplots()
ax.plot(T-273.15,k1,'r*-',label="k1")
ax.plot(T-273.15,k2,label="k2")
plt.xlabel('T(oC)')
ax.legend(loc=2);
```

```python
keq=k1/k2
```

```python
plt.plot(T-273.15,keq)
plt.xlabel('T (oC)')
plt.ylabel('Keq')
```
